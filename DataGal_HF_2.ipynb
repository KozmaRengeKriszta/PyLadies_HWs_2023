```python
# 'eletkor' oszlop létrehozása
nepszamlalas['eletkor'] = 2023 - nepszamlalas['szuletesi_ev']

#szépkorúak (110+ év) szűrése és külön DF-be mentése
szepkoruak = nepszamlalas[nepszamlalas.eletkor >= 110].copy(deep=True)
nem_szepkoruak = nepszamlalas[nepszamlalas.eletkor < 110].copy(deep=True)
```

```python
#hány lakos töltötte már be legalább a 110. életévét? És mennyien vannak a többiek?

szepkoruak.shape 
# 1177 bergengóc töltötte már be a 110. életévét
nem_szepkoruak.shape 
# 9349-en vannak a többiek

# alternatív megoldás a .count() függvény használatával
szepkoruak.ID.count()
nem_szepkoruak.ID.count()

# ellenőrzés: 1177 + 9349 = 10526, ami a teljes minta
```

```python
# egyik oszlopban fura értékek voltak - gyűjtsük ki a hibásnak tűnő értékeket

nepszamlalas['magassag'].sort_values(ascending = False) 
# a 220 cm-es magasság tűnik még reálisnak, az ennél magasabb értékeket hibásnak vélem

# iteráció segítségével a következőképpen oldanám meg a feladatot:
for index, sor in nepszamlalas.iterrows():
    if sor['magassag'] > 220:
        print(f"Az {index}. sorban található magasság értéke irreális: {sor['magassag']} cm")

# alternatív megoldás iteráció nélkül:
nepszamlalas[nepszamlalas.magassag > 220]

# hibás magasság értékek: 1030, 1180, 1280, 1690, 1850
```

```python
#népcsoportokat takaró kódszámok: 0 - ber, 1 - gen, 2 - goc, 3 - bevandorlo >> dekódolt táblázat létrehozása
for index, sor in nepszamlalas.iterrows():
    if sor['nepcsoport'] == 0:
        nepszamlalas.loc[index, 'nepcsoport'] = 'ber'
    elif sor['nepcsoport'] == 1:
        nepszamlalas.loc[index, 'nepcsoport'] = 'gen'
    elif sor['nepcsoport'] == 2:
        nepszamlalas.loc[index, 'nepcsoport'] = 'goc'
    else:
        nepszamlalas.loc[index, 'nepcsoport'] = 'bevandorlo'

nepszamlalas
        
# első alternatív megoldás dataframe szűrést használva
    #nepszamlalas.loc[nepszamlalas['nepcsoport'] == 0, 'nepcsoport'] = 'ber'
    #nepszamlalas.loc[nepszamlalas['nepcsoport'] == 1, 'nepcsoport'] = 'gen'
    #nepszamlalas.loc[nepszamlalas['nepcsoport'] == 2, 'nepcsoport'] = 'goc'
    #nepszamlalas.loc[nepszamlalas['nepcsoport'] == 3, 'nepcsoport'] = 'bevandorlo'


# második alternatív megoldás a .replace() függvény használatával
    #nepszamlalas['nepcsoport'] = nepszamlalas['nepcsoport'].replace({0: 'ber', 1: 'gen', 2: 'goc', 3:'bevandorlo'})
```

```python
# Hány "dzs" és "x" fordul elő a Bergengóc Himnusz óbergengóc változatában?

himnusz = "A nexdzsebdzsal itek mvip, zohüs ó nexdzsebdzsal bip,\
          Hüxés nytlm coxütöubc ótcékkó éxmvüdzsubc\
          Bexdzsebdzsaloó he ja!\
          Nes cett sémk ódzsdzsahbubc süx, xübc ó cybböenn itek züx\
          nytlm, exőm im nükéx, edzsö ó nexdzsebdzsal künéx\
          Bexdzsebdzsaloó, ke ja! Bexdzsebdzsaloó, dóda!"

dzs_szamlalo = 0
x_szamlalo = 0

# végig iterálok a szövegen és ha 'dzs' string szeletet találok, akkor növelem a dzs_szamlalo értékét 1-gyel
# a 3 hosszúságú 'dzs' szelet miatt a teljes szöveghossz - 2 hosszúságban, hogy az himnusz[i:i+3] ne fusson ki a teljes szöveghosszból

for i in range(0, len(himnusz) - 2):
    if himnusz[i:i+3] == 'dzs':
        dzs_szamlalo += 1

# végig iterálok a szövegen és ha 'x' betűt találok, akkor növelem az x_szamlalo értékét 1-gyel
for betu in himnusz:
    if betu == 'x':
        x_szamlalo += 1

print('dzs = ' + str(dzs_szamlalo) + ', ' + 'x = ' + str(x_szamlalo))

# A 'dzs' betű 16-szor, az 'x' betű 15-ször fordul elő a Bergengóc Himnusz óbergengóc változatában
```
